```python
daily_sales = df.groupby('Order Date')['Sales'].sum().reset_index()
```

```python
monthly_sales = df.resample('ME', on='Order Date')['Sales'].sum().reset_index()
```

```python
df['Year'] = df['Order Date'].dt.year
yearly_sales = df.groupby('Year')['Sales'].sum().reset_index()
```

```python
plt.figure(figsize=(8,5))
plt.plot(monthly_sales['Order Date'], monthly_sales['Sales'],linewidth = 2)
plt.title('Sales over time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid()
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Day'] = df['Order Date'].dt.day
df['Weekday'] = df['Order Date'].dt.day_name()
```

```python
weekday_sales = df.groupby('Weekday')['Sales'].sum().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
]).reset_index()

plt.figure(figsize=(10,5))
sns.barplot(x='Weekday', y='Sales', data=weekday_sales)
plt.title("Weekday Sales Pattern")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum().reset_index()

plt.figure(figsize=(10,5))
sns.lineplot(x='Month', y='Sales', data=monthly_sales, marker='o')
plt.title("Monthly Sales Seasonality")
plt.xticks(range(1,13))
plt.grid(True)
plt.show()
```

```python
# Select numeric columns only
numeric_df = df[['Sales', 'Profit', 'Quantity', 'Discount']]

# Correlation matrix
corr_matrix = numeric_df.corr()
print(corr_matrix)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(numeric_df)
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```
